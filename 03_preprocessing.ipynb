```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data (not raw!)
df = pd.read_csv("02_EDA .csv")
df.head()
```

```python
# AgeGroup was created by us for EDA only
# Model doesn't need it since Age column already exists
df = df.drop("AgeGroup", axis=1)
print(df.columns.tolist())
```

```python
# Convert Geography and Gender from text to numbers
df = pd.get_dummies(df, columns=["Geography", "Gender"], drop_first=True)
print(df.columns.tolist())
print(df.head())
```

```python
# X = everything model uses to predict
# y = what model is trying to predict

X = df.drop("Exited", axis=1)
y = df["Exited"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```
