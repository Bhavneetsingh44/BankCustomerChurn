```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data (not raw!)
df = pd.read_csv("03_preprocessing .csv")
df.head()
```

```python
# How much of their salary do they keep in bank?
df["BalanceToSalary"] = (df["Balance"]) / (df["EstimatedSalary"] + 1)

# +1 avoids division by zero
# High ratio = keeps more money in bank = more invested?
df["BalanceToSalary"].head()
```

```python
# Are they using many products relative to how long they've been here?
df["ProductsPerYear"] = df["NumOfProducts"] / (df["Tenure"] + 1)

# +1 avoids division by zero for new customers

df["ProductsPerYear"].head()
```

```python
# Active member with more products = truly engaged?
df["EngagementScore"] = df["IsActiveMember"] * df["NumOfProducts"]

# Active + 2 products = score of 2
# Inactive + 2 products = score of 0
df["EngagementScore"].head()
```

```python
# How old were they when they joined?
df["AgeWhenJoined"] = df["Age"] - df["Tenure"]

# Older customer who just joined is very different
# from young customer who joined young and stayed

df["AgeWhenJoined"].head()
```
